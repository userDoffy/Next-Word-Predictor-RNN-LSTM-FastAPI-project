```python
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
import pandas as pd
import numpy as np

df=pd.read_csv("datasets/cleaned.csv")
df.head()
sents=df['text'].tolist()
sents[:4]
```

Tokenizing the each sentences into numbers

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts(sents)
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
[tokenizer.texts_to_sequences([t])[0] for t in sents][:5]
```

Creating a Input Output Sequence for Training

```python
input_sequences=[]
for t in sents:
    tokenized_sent=tokenizer.texts_to_sequences([t])[0]

    for i in range(1,len(tokenized_sent)):
        input_sequences.append(tokenized_sent[:i+1])
input_sequences
```

Padding

```python
max_len=max([len(x) for x in input_sequences])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences=pad_sequences(input_sequences,maxlen=max_len,padding='pre')
padded_input_sequences
```

Let us consider all the texts before the last one will be our output

```python
inputs=padded_input_sequences[:,:-1]
inputs
```

```python
outputs=padded_input_sequences[:,-1]
outputs
```

```python
inputs.shape,outputs.shape
```

Converting Outputs to Categorical Classification

```python
num_classes=len(tokenizer.word_index)+1
num_classes
```

Rest of Stuff Done on Cloud for wikitext dataset

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(outputs,num_classes=num_classes)
y,y.shape
```

LSTM Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense,GRU
from tensorflow.keras.layers import Bidirectional
```

```python
model = Sequential()
model.add(Embedding(num_classes, 100, input_length=max_len-1))
model.add(Bidirectional(LSTM(150)))
model.add(Dense(num_classes, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
model.fit(inputs,y,epochs=50) 
```

```python
text="Tommorow i" 
```

```python
#tokenize
token_text=tokenizer.texts_to_sequences([text])[0]
#padding
padded_token_text=pad_sequences([token_text],maxlen=max_len-1,padding='pre')
#predict
output=model.predict(padded_token_text)
ind=np.argmax(output)
for word,index in tokenizer.word_index.items():
    if index==ind:
        text=text+" "+word
        break
text
```
